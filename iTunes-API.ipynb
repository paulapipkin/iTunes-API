### Itunes API

- API review

info= https://developer.apple.com/library/archive/documentation/AudioVideo/Conceptual/iTuneSearchAPI/Searching.html#//apple_ref/doc/uid/TP40017632-CH5-SW1

```python
import requests
import json
```

```python
base_site = 'https://itunes.apple.com/search'
```

```python
url = base_site + "?term=the+beatles&country=us"

requests.get(url)
```

```python
#using requests.get dictionary

r = requests.get(base_site, params = {'term':'the beatles', 'country':'us'})
r.status_code
```

```python
r.url
```

```python
#Checking outputs

info = r.json()
print(json.dumps(info, indent=4))
```

```python
info.keys()
```

```python
print(json.dumps(info['results'][0], indent=4))
```

```python
info['resultCount']
```

```python
tracks = (info['results'][1]['artistName'])
tracks
```

```python
#grabbing information
tracks = []
for n in range(0,info['resultCount']):
    artist_name = info['results'][n]['artistName']
    tracks.append(artist_name)
print(tracks)
```

```python
## Getting all names and release dates for 200 results
```

```python
r = requests.get(base_site, params = {'term':'the beatles', 'country':'us', 'limit':200})
r.ok
```

```python
type(r)
```

```python
info = r.json()
info.keys()
```

```python
print(json.dumps(info['results'][45]['trackName'], indent=4))
```

```python
#grabbing information - Names and Dates
dict_name_date = {'Name':[],'Release Date':[]}
for result in info['results']:
    
    dict_name_date["Release Date"].append(result['releaseDate'])
    dict_name_date["Name"].append(result['artistName'])
print(json.dumps(dict_name_date, indent=4))
```

```python
#alternative:

for result in info['results']:
    print(result['artistName'])
```

```python
#Structuring and exporting data
import pandas as pd
df = pd.DataFrame(info['results'])
df.head()
```
